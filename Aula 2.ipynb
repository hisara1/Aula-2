```python
import math
#math.modf(x)  (0.5678000000000338, 1234.0)

print('A base em binário ', bin(10))
r_1 = 2**-4 + 2**-5 + 2**-6
print( 'Valor em binário a ser convertindo é r_1 = ', r_1 )

print('\nPrimeira iteração...\n')
w1 = r_1 * 10
print('\nw1 = r_1 * 10 = ', w1)

fracInt = math.modf(w1)
b_1 = fracInt[1]
r_2 = fracInt[0]
print('Convensão de z1 (parte inteira) para a base 10,', b_1)
print('A parte fracionária do número binário é r_2 =',r_2)

print('\nSegunda iteração...\n')

w2 = r_2 * 10
print('\nw2 = r_2 * 10', w2)

fracInt = math.modf(w2)
b_2 = fracInt[1]
r_3 = fracInt[0]
print('Convensão de z2 (parte inteira) para a base 10,', b_2) 
print('A parte fracionária do número binário é r_3 =',r_3)

print('\nTerceira iteração...\n')

w3 = r_3 * 10
print('\nw3 = r_3 * 10', w3)

fracInt = math.modf(w3)
b_3 = fracInt[1]
r_4 = fracInt[0]
print('Convensão de z2 (parte inteira) para a base 10,', b_3) 
print('A parte fracionária do número binário é r_3 =',r_4)
```

```python
num = str(0.00153)
base = 10
dig = 4
exps = 5
expi = -5
def truncar (num,base,dig,exps,expi):
    i = 0
    k = 0
    trunc = ""
    L = []
    a = num.split(".")
    p0 = a[0]

    exp = len(p0)
    aux = float(num)

    aux = aux*base**-exp
    numt = len(str(aux)) - 1
    aux = str(aux)
    if int(exp) > exps:
      print("OverflowError")
    elif exp < int(expi):
      print("UnderflowError")

    while i <= numt:
      L.append(aux[i])
      i = i + 1

    del (L[2+dig:numt+1])
   
    while k < len(L):
      trunc = trunc + L[k]
      k = k + 1

    if(int(p0)>0):
      print("O valor truncado e: " + trunc+ "*10**"+str(exp))
    else:
      print("O valor truncado e: " + trunc+ "*10**"+'1')
    
truncar(num,base,dig,exps,expi)
```
